## Problem 1

````python
import numpy as np
import matplotlib.pyplot as plt

def beauty_derivative_plot(x0):
    f = lambda x: 2 * np.sin(x) - np.cos(2 * x)
    df = lambda x: 2 * np.cos(x) + 2 * np.sin(2 * x)


    x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
    y = f(x)
    y_der = df(x)

    y0 = f(x0)
    slope = df(x0)

    x_tan = np.linspace(x0 - 1, x0 + 1, 200)
    y_tan = y0 + slope * (x_tan - x0)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label=r"$f(x)=2\sin x-\cos 2x$", linewidth=2)
    plt.plot(x, y_der, "--", label=r"$f'(x)=2\cos x+2\sin 2x$", linewidth=2)
    plt.plot(x_tan, y_tan, label=f"Tangent at x = {x0:.2f}", linewidth=2)

    plt.scatter([x0], [y0], s=60, zorder=5)


    plt.annotate(
        f"({x0:.2f}, {y0:.2f})",
        xy=(x0, y0),
        xytext=(x0 - 1.0, y0 - 1.0)
    )

    plt.xlim(-2 * np.pi, 2 * np.pi)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Function, Derivative, and Tangent Line")
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.show()
beauty_derivative_plot(-3)
beauty_derivative_plot(0)
beauty_derivative_plot(3)

````
